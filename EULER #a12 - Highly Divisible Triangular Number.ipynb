```python
'''This is a MUCH faster method of finding the divisors; I remodeled in excel to make sure I understand...'''
'''E.g. 36: Check "1" 36 / 1 has no remainder so add 1 to list. 
''''''Since 1 doesnt equal 36 / 1 IT ALSO APPENDS 36 / 1 to list AKA 36. Check 2 it adds 2 and 18, Check 6 it would only add 6 because 36 / 6 = 6'''
'''With this method we are quickly adding twice as many numbers very quickly and not iterating through the whole list for each number...'''
from math import *
def divisor(n):
    boundary = int(sqrt(n))
    divisor_list = []
    for i in range(1, boundary + 1, 1):
        if n % i == 0:
            divisor_list.append(i)
            if i != n/i:
                divisor_list.append(n/i)
    return len(divisor_list)
```
